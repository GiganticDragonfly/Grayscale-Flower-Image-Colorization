# B&W Flowers Image Colorization - Testing the trained model

After we trained the model, it's time to test it.<br>
The code below loads images paths from a given folder,<br>
and outputs a prediction for each image using the trained model.<br>

```python
import keras
import numpy as np
import matplotlib.pyplot as plt
from keras.models import load_model,Model
from PIL import Image
import glob as glob
```

```python
IMAGE_DIR= './test/' #Path to image folder to predicte
```

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])
```

```python
model = load_model('./models/d_model.h5py')
```

```python
images_names= glob.glob(IMAGE_DIR+"*")
images = []
print()

for name in images_names:
    image = Image.open(name)
    image = np.array(image,dtype=np.float32)
    image = rgb2gray(image)
    image = image/255
  
    images.append(image)

images = np.array(images)
images = images.reshape(images.shape[0],32,32,1)

images = np.array(images[:])
prediction = model.predict(images)

prediction = np.array(prediction)
for i in range (len(prediction)):
    plt.imshow(images[i].reshape(32,32),cmap='gist_gray')
    plt.show()
    plt.imshow(prediction[i])
    plt.show()
    print("----------------------------")
```
